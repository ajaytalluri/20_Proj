```python
# Split data in to x and y
x=df.drop(columns=["Attrition"])
y=df[["Attrition"]]
```

```python
for index in x.skew().index:
    if x.skew().loc[index]>0.5:
        x[index]=np.log1p(x[index])
    if x.skew().loc[index]<-0.5:
        x[index]=np.log1p(x[index])
```

```python
print(x.skew())
print("Total no of numeric features: ",len(x.skew()))
print("No of Features which are skewed:  ", len(x.skew().loc[abs(x.skew())>0.5]))
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,accuracy_score
def max_auc_roc_score(algo,x,y):
    max_auc_roc_score=0
    for r_state in range(42,110):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r_state,test_size=0.20)
        algo.fit(x_train,y_train)
        y_pred = algo.predict(x_test)
        aucroc_scr=roc_auc_score(y_test,y_pred)
        print("auc roc score corresponding to ",r_state," is ",aucroc_scr)
        if aucroc_scr>max_auc_roc_score:
            max_auc_roc_score=aucroc_scr
            final_r_state=r_state
    print("max auc roc score corresponding to ",final_r_state," is ",max_auc_roc_score)
    print("Accuracy is=",accuracy_score(y_test,y_pred))
    return final_r_state
```

```python
from sklearn.linear_model import LogisticRegression
algo=LogisticRegression()
max_auc_roc_score(algo,x,y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(algo,x,y,cv=5,scoring="roc_auc").mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
algo=DecisionTreeClassifier()
max_auc_roc_score(algo,x,y)
```

```python
cross_val_score(algo,x,y,cv=5,scoring="roc_auc").mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
parameters={"n_estimators":[10,100,500]}
rf=RandomForestClassifier()
gsv = GridSearchCV(rf, parameters, cv=5,scoring="roc_auc")
gsv.fit(x,y)
gsv.best_params_
```

```python
algo =RandomForestClassifier(n_estimators=500)
max_auc_roc_score(algo,x,y)
```
