```python
# let us create new features to find the penetration 
df['test_per_pop_perc']=(df['Total Tests']/df['Population_y'])*100
```

```python
df.head()
```

```python
#let us call 0-14 years as Kid, 15-65 adult and 65+ old
df['pop_kid']=df['Population_y']*df['Fraction age 0-14 years']
df['pop_adult']=df['Population_y']*df['Fraction age 15-64 years']
df['pop_old']=df['Population_y']*df['Fraction age 65+ years']
```

```python
df
```

```python
df['Deat_per_pop_per']=(df['Total Deaths']/df['Population_y'])*100
```

```python
df['Deat_per_pop']=df['Total Deaths']/df['Population_y']
```

```python
df
```

```python
# probable deaths from kids
df['prob_kids']=df['Deat_per_pop']*df['pop_kid']
```

```python
df['prob_adult']=df['Deat_per_pop']*df['pop_adult']
df['prob_old']=df['Deat_per_pop']*df['pop_old']
```

```python
df
```

```python
df.fillna(0)
```

```python
aa=df['Country'].groupby(df.prob_adult.nlargest(20))
#plt.show()
```
